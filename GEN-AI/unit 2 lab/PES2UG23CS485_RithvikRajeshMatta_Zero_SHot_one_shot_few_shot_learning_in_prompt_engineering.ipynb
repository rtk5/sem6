```python
from transformers import pipeline
```

BART = encoder–decoder Transformer

MNLI = trained on Natural Language Inference

Not trained for sentiment or topic classification

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)
```

```python
text = "The course content was informative but very fast paced."

labels = ["positive", "negative", "neutral"]
```

```python
result = classifier(text, labels)

print(result)
```

```python
from transformers import pipeline

classifier = pipeline(
    "text-generation",   # NOT any-to-any
    model="google/flan-t5-small"
)

result = classifier("Explain quantum computing in simple terms.")
print(result)
```

```python
prompt = """
Classify the sentiment of the feedback.

Example:
Feedback: "The class was very interesting"
Sentiment: Positive

Now classify:
Feedback: "The lab sessions were confusing"
Sentiment:
"""
```

```python
result = classifier(prompt, max_new_tokens=10)
print(result[0]["generated_text"])
```

```python
#few shot learrning
from transformers import pipeline

classifier = pipeline(
    "text-generation",
    model="google/flan-t5-small"
)
```

```python
prompt = """
Classify the sentiment of the feedback.

Examples:
Feedback: "The class was very interesting"
Sentiment: Positive

Feedback: "The instructor explained concepts clearly"
Sentiment: Positive

Feedback: "The lab sessions were boring"
Sentiment: Negative

Now classify:
Feedback: "The lab sessions were confusing"
Sentiment:
"""
```

```python
result = classifier(prompt, max_new_tokens=10)
print(result[0]["generated_text"])
```

| Type      | Examples in Prompt |
